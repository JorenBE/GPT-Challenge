```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
FILENAME = 'ClickActivationE.csv'
```

```python
df = pd.read_csv(FILENAME, index_col=0)
print(len(df))
```

```python
df['E_bin'] = pd.qcut(df['Free energy of activation (kcal/mol)'], 2, labels=[0,1])
df['smiles_lenght'] = [len(list(i)) for i in df['SMILES'].values]
```

```python
df['Free energy of activation (kcal/mol)'].hist()
plt.xlabel('Free energy of activation (kcal/mol)')
```

```python
plt.scatter(df['smiles_lenght'] , df['Free energy of activation (kcal/mol)'], c= df['E_bin'])
plt.hlines(df['Free energy of activation (kcal/mol)'].median(), 0, 50, colors= 'maroon')
```
