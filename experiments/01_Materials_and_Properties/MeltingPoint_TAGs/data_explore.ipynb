```python
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
FILE = 'dataset_TAG_full.csv'
df = pd.read_csv(os.path.join('data',FILE ), delimiter= ';')
print(df.columns)
```

```python
df.dropna(subset='[C ]', inplace=True)
df['mp_bin'] = pd.qcut(df['[C ]'], 2, labels=[0,1])
```

```python
fig, ax = plt.subplots(1,2)
ax  = ax.flatten()
df['[C ]'].hist(ax = ax[0])
ax[0].vlines(df['[C ]'].median(), 0, 50, colors = 'black')
print(df['[C ]'].median())
ax[1].pie([df['mp_bin'].tolist().count(0), df['mp_bin'].tolist().count(1)],labels = [0,1])
```

```python
df.to_csv('fats_data.csv')
```
