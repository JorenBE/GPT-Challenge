```python
from fastcore.xtras import load_pickle
import pandas as pd 
import os 
from glob import glob
import matplotlib.pyplot as plt
from datetime import datetime
import ast

from gptchem_llama.analyze import OutPickle, OutFolder, add_traditional_ml
import gptchem_llama.plotting

from gptchem_llama.plotting import *

FOLDER_LLAMA = 'LearningCurve_models/out_llama'
FOLDER_GPTJ = 'LearningCurve_models/out_gptj'
FOLDER_MISTRAL = 'LearningCurve_models/out_mistral'
```

```python
print('Llama')
try:
    folder_llma = OutFolder(FOLDER_LLAMA)
    folder_llma_df = folder_llma.get_df(df_type= 'analysis', exclude = {'train_size': [5]})
    display(folder_llma_df)
except:
    print('Data not present')
```

```python
print('Mistral')
try:
    folder_mistral = OutFolder(FOLDER_MISTRAL)
    folder_mistral_df = folder_mistral.get_df(df_type= 'analysis', exclude = {'train_size': [5]})
    display(folder_mistral_df)
except:
    print('Data not present')
```

```python
print('GPTJ')
try:
    folder_gptj = OutFolder(FOLDER_GPTJ)
    folder_gptj_df = folder_gptj.get_df(df_type= 'analysis', exclude = {'train_size': [5]})
    display(folder_gptj_df)
except:
    print('Data not present')
```

```python
def plot():
    fig, ax = plt.subplots(4,1, figsize=(ONE_COL_WIDTH_INCH, TWO_COL_GOLDEN_RATIO_HEIGHT_INCH),constrained_layout = True , sharex=True)
    try:
        folder_llma.plot_analysis(bins = 2, representation='reaction', target = 'yield_bin', ax = ax, exclude = {'train_size': [5]})
        folder_gptj.plot_analysis(bins = 2, representation='reaction', target = 'yield_bin', ax = ax, exclude = {'train_size': [5]})
        folder_mistral.plot_analysis(bins = 2, representation='reaction', target = 'yield_bin', ax = ax, exclude = {'train_size': [5]})

        folder_mistral.add_zero_rule(ax = ax, color = 'slategray', exclude = {'train_size': [5]})
    except:
        print('No data yet')
    ax = add_traditional_ml(ax, file = './ml/Yield_isomerisation_traditional_ml.csv', target='yield_bin')
    #ax = add_zero_rule(ax)

    import matplotx
    matplotx.line_labels(ax[0])

    ax[3].tick_params(axis='x', rotation=45)

    fig.savefig('Isomerization_LearningCurves_allModels.pdf')

plot()

```
