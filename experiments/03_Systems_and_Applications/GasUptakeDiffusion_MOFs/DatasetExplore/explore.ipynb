```python
import pandas as pd
import matplotlib.pyplot as plt
import sys
sys.path.append("/home/jorenvanherck/Documents/gptchallengeV2/gptchem-gptj/plotutils/")

plt.style.use("/home/jorenvanherck/Documents/gptchallengeV2/gptchem-gptj/plotutils/kevin.mplstyle")
```

```python
fig, axes = plt.subplots(4, 2, figsize = (6,6), tight_layout = True, 
                         )
files = {"He": 'Helium.csv',
         "H2": "Hydrogen_data.csv",
         "CH4": "Methane.csv",
         "N2" : "N2_data.csv"}
property = {'Uptake': 'U_{element} (mol/kg)',
            'Diffusion': 'D_{element} (cm2/s)'}

for i, (name, file) in enumerate(files.items()):
    df = pd.read_csv(file, index_col=0)

    for j, (prop, col) in enumerate(property.items()):
        df[col.format(element = name)].hist(ax =axes[i][j], density = False)
        axes[i][j].set_xlabel(f'{name} {prop}')

axes[0][0].set_xlabel(r"$He$" + " Uptake")
axes[0][1].set_xlabel(r"$He$" + " Diffusion")

axes[1][0].set_xlabel(r"$H_2$" + " Uptake")
axes[1][1].set_xlabel(r"$H_2$" + " Diffusion")

axes[2][0].set_xlabel(r"$CH_4$" + " Uptake")
axes[2][1].set_xlabel(r"$CH_4$" + " Diffusion")

axes[3][0].set_xlabel(r"$N_2$" + " Uptake")
axes[3][1].set_xlabel(r"$N_2$" + " Diffusion")

axes[2][1].set_xticks([0.00020, 0.00050, 0.00080])

plt.savefig('MOF_uptake_histograms.pdf',bbox_inches='tight' )
```
